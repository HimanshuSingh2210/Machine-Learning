```python
import numpy as np
import pandas as pd
```

```python

df = pd.read_csv("/content/Cricket.csv",encoding='latin1')
```

```python
df
```

```python
df.info()
```

```python
#replacing * sign from the data
df['HS']=df.HS.str.replace('*','')
```

```python
df.info()
```

```python
#changing datatype
df["HS"]=df.HS.astype(int)
```

```python
df.describe().T
```

```python
# null values
df.isnull().sum()
```

```python
# to convert the span into years of experience
# we first split the span into start and end and store in new column

df[['Strt','End']] = df.Span.str.split("-",expand=True)
```

```python
df.head()
```

```python
# convert them as int and
# find years of exp as end year - start year
# later drop the unneceassry columns such as start, end and span
# we are only left with the experience column now

df[['Strt','End']]=df[['Strt','End']].astype(int)
df['Exp']=df['End']-df['Strt']
```

```python
df.describe().T
```

```python
#droping strt and end column because we already calculated total exp
df=df.drop(['Strt','End','Span'], axis = 1)
df.head()
```

```python
# Dropping Player field as final dataframe will only contain data columns after copying

df_drop = df.copy()
player = df_drop.pop('Player')
```

```python
# Rescaling the Features/standardising
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_drop)
df_scaled
```

```python
df_df1 = pd.DataFrame(df_scaled, columns = [ 'Mat', 'Inns', 'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100',
                                            '50', '0', 'Exp'])
df_df1.head()
```

```python
df_df1.describe().T
```

```python
# Model Building
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
```

```python
# mergings
#metric is calculated by euclidean distance
#method-their are several method but we are using complete here
mergings = linkage(df_df1, method='complete',metric='euclidean')
```

```python
mergings
```

```python
# Complete Linkage and representing in form of dendogram
plt.figure(figsize = (20,10))
mergings = linkage(df_df1, method='complete',metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
from scipy.cluster.hierarchy import cut_tree
```

```python
#as we can see that 2 cluster is best shown in previous block but clustering is a subjective call we are trying with 4 and 3 cluster for more clearification
# 4 clusters
#reshape changes horizontal to vertical

cluster_labels = cut_tree(mergings, n_clusters=4).reshape(-1, )
cluster_labels
```

```python
# 3 clusters
cluster_labels3 = cut_tree(mergings, n_clusters=3).reshape(-1, )
cluster_labels3
```

```python
# assign cluster labels
df_final = df.copy()
df_final.head()
```

```python
df_final['Cluster_labels_4'] = cluster_labels
df_final['Cluster_labels_3'] = cluster_labels3
df_final.head()
```
